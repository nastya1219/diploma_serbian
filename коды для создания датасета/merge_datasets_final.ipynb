# Merge Two Serbian Verb Datasets

This notebook combines two verb dataset files into a single merged file.

## 1) Import and Configure

```python
import os
import pandas as pd

notebooks_dir = os.getcwd()
file1_path = os.path.join(notebooks_dir, 'data', 'csv', 'serbian_verb_final_hf_wiki_merged.csv')
file2_path = r'C:\Users\Анастасия Андреевна\Downloads\диплом_4 курс\data\csv\serbian_verb_final_02_02_with_ruscorpora.csv'
output_path = os.path.join(notebooks_dir, 'data', 'csv', 'serbian_verb_final_combined.csv')

print(f'File 1: {file1_path}')
print(f'File 2: {file2_path}')
print(f'Output: {output_path}')
```

## 2) Check Files

```python
file1_exists = os.path.exists(file1_path)
file2_exists = os.path.exists(file2_path)

print(f'File 1 exists: {file1_exists}')
print(f'File 2 exists: {file2_exists}')

if file1_exists and file2_exists:
    print('✓ Both files found!')
```

## 3) Merge Datasets

```python
if file1_exists and file2_exists:
    # Load datasets
    df1 = pd.read_csv(file1_path)
    df2 = pd.read_csv(file2_path)
    
    print(f'File 1: {len(df1)} rows')
    print(f'File 2: {len(df2)} rows')
    print(f'Columns: {list(df1.columns)}\n')
    
    # Combine both datasets
    merged = pd.concat([df1, df2], ignore_index=True)
    print(f'Combined (before aggregation): {len(merged)} rows')
    
    # Group by (base_verb, prefixed_verb, prefix) and sum frequencies
    merged_unique = merged.groupby(['base_verb', 'prefixed_verb', 'prefix'], as_index=False).agg({
        'freq_base': 'sum',
        'freq_prefixed': 'sum'
    })
    
    print(f'Final (after aggregation): {len(merged_unique)} rows')
    print(f'\nExample aggregation (dobiti):')
    dobiti_rows = merged_unique[merged_unique['prefixed_verb'] == 'dobiti']
    if len(dobiti_rows) > 0:
        print(dobiti_rows.head())
    
    # Save result
    merged_unique.to_csv(output_path, index=False, encoding='utf-8-sig')
    print(f'\n✓ Saved: {output_path}')
else:
    print('❌ Cannot merge - files not found')
```

## 4) View Results

```python
if file1_exists and file2_exists and 'merged_unique' in locals():
    print('First 10 rows of merged dataset:')
    print(merged_unique.head(10))
    print(f'\nTotal unique verb pairs: {len(merged_unique)}')
```
