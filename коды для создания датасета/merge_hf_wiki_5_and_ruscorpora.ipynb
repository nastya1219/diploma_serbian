# Merge Aggregated HF Wiki 5 with RusCorpora Data

Combines `serbian_verb_final_hf_wiki_5_merged_aggregated.csv` with `serbian_verb_final_02_02_with_ruscorpora.csv` and sums frequencies.

## 1) Import and Configure Paths

```python
import os
import pandas as pd

notebooks_dir = os.getcwd()

# File 1: Aggregated HF Wiki 5 (from merge_hf_wiki_5_aggregated notebook)
file1_path = os.path.join(notebooks_dir, 'data', 'csv', 'serbian_verb_final_hf_wiki_5_merged_aggregated.csv')

# File 2: RusCorpora data
file2_path = r'C:\Users\Анастасия Андреевна\Downloads\диплом_4 курс\data\csv\serbian_verb_final_02_02_with_ruscorpora.csv'

# Output file
output_path = os.path.join(notebooks_dir, 'data', 'csv', 'serbian_verb_final_hf_wiki_5_ruscorpora_combined.csv')

print(f'File 1: {file1_path}')
print(f'File 2: {file2_path}')
print(f'Output: {output_path}')
```

## 2) Check Files Existence

```python
file1_exists = os.path.exists(file1_path)
file2_exists = os.path.exists(file2_path)

print(f'File 1 exists: {file1_exists}')
print(f'File 2 exists: {file2_exists}')

if file1_exists and file2_exists:
    print('\n✓ Both files found! Ready to merge.')
else:
    if not file1_exists:
        print(f'\n⚠️ File 1 not found: {os.path.basename(file1_path)}')
    if not file2_exists:
        print(f'\n⚠️ File 2 not found: {os.path.basename(file2_path)}')
```

## 3) Load and Merge with Frequency Aggregation

```python
if file1_exists and file2_exists:
    # Load datasets
    df1 = pd.read_csv(file1_path)
    df2 = pd.read_csv(file2_path)
    
    print(f'File 1: {len(df1)} rows')
    print(f'File 2: {len(df2)} rows')
    print(f'Columns: {list(df1.columns)}\n')
    
    # Combine both datasets
    merged = pd.concat([df1, df2], ignore_index=True)
    print(f'Combined (before aggregation): {len(merged)} rows')
    
    # Group by (base_verb, prefixed_verb, prefix) and sum frequencies
    merged_agg = merged.groupby(['base_verb', 'prefixed_verb', 'prefix'], as_index=False).agg({
        'freq_base': 'sum',
        'freq_prefixed': 'sum'
    })
    
    print(f'Final (after aggregation): {len(merged_agg)} rows')
    print(f'Duplicates aggregated: {len(merged) - len(merged_agg)}')
    
    print(f'\nExample aggregation (biti):') 
    biti_rows = merged_agg[merged_agg['base_verb'] == 'biti']
    if len(biti_rows) > 0:
        print(biti_rows.head())
    
    # Save result
    merged_agg.to_csv(output_path, index=False, encoding='utf-8-sig')
    print(f'\n✓ Saved: {output_path}')
else:
    print('❌ Cannot merge - one or both files not found')
```

## 4) View Results

```python
if file1_exists and file2_exists and 'merged_agg' in locals():
    print('First 10 rows of final merged dataset:')
    print(merged_agg.head(10))
    print(f'\nTotal unique verb pairs: {len(merged_agg)}')
```
