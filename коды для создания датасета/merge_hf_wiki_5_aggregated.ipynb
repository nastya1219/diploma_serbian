# Merge All HF Wiki Verb Datasets with Frequency Aggregation

Combines all `serbian_verb_final_hf_wiki_5_*.csv` files and sums frequencies for duplicate verb pairs.

## 1) Import and Configure

```python
import os
import pandas as pd
import glob

notebooks_dir = os.getcwd()
data_dir = os.path.join(notebooks_dir, 'data', 'csv')

# Find all HF Wiki files
pattern = os.path.join(data_dir, 'serbian_verb_final_hf_wiki_5_*.csv')
files = sorted(glob.glob(pattern))

output_path = os.path.join(data_dir, 'serbian_verb_final_hf_wiki_5_merged_aggregated.csv')

print(f'Data directory: {data_dir}')
print(f'Output file: {output_path}')
```

## 2) List Files to Merge

```python
print(f'Found {len(files)} files:\n')
for f in files:
    print(f'  - {os.path.basename(f)}')

if len(files) == 0:
    print('\n❌ No files found!')
```

## 3) Merge and Aggregate

```python
if len(files) > 0:
    # Load all files
    dfs = [pd.read_csv(f) for f in files]
    
    # Show info about each file
    print('File sizes:')
    for f, df in zip(files, dfs):
        print(f'  {os.path.basename(f)}: {len(df)} rows')
    
    # Combine all datasets
    merged = pd.concat(dfs, ignore_index=True)
    print(f'\nCombined (before aggregation): {len(merged)} rows')
    
    # Group by (base_verb, prefixed_verb, prefix) and sum frequencies
    merged_agg = merged.groupby(['base_verb', 'prefixed_verb', 'prefix'], as_index=False).agg({
        'freq_base': 'sum',
        'freq_prefixed': 'sum'
    })
    
    print(f'Final (after aggregation): {len(merged_agg)} rows')
    print(f'\nDuplicates removed: {len(merged) - len(merged_agg)}')
    
    # Save result
    merged_agg.to_csv(output_path, index=False, encoding='utf-8-sig')
    print(f'\n✓ Saved: {output_path}')
else:
    print('❌ Cannot merge - no files found')
```

## 4) View Results

```python
if len(files) > 0 and 'merged_agg' in locals():
    print('First 10 rows of merged aggregated dataset:')
    print(merged_agg.head(10))
    print(f'\nColumns: {list(merged_agg.columns)}')
    print(f'Total unique verb pairs: {len(merged_agg)}')
```
