# Merge Serbian HF Wiki Verb Datasets

This notebook combines all `serbian_verb_final_hf_wiki_5_*.csv` files into a single merged dataset with deduplication.

## 1) Import Required Libraries

```python
import os
import pandas as pd
import glob

# Set output directory
OUT_DIR = os.path.join(os.getcwd(), "data", "csv")
```

## 2) Find and List Files

```python
# Найти все файлы с паттерном serbian_verb_final_hf_wiki_5_*.csv
pattern = os.path.join(OUT_DIR, "serbian_verb_final_hf_wiki_5_*.csv")
files = sorted(glob.glob(pattern))

print(f"Найдено файлов: {len(files)}")
for f in files:
    print(f"  - {os.path.basename(f)}")
```

## 3) Merge All Files

```python
# Прочитать и объединить все файлы
if files:
    dfs = [pd.read_csv(f) for f in files]
    merged_df = pd.concat(dfs, ignore_index=True)
    
    print(f"\nВсего строк в объединённом датасете: {len(merged_df)}")
    print(f"Колонки: {list(merged_df.columns)}")
    
    # Удалить дубликаты если есть
    merged_df_unique = merged_df.drop_duplicates(subset=['prefixed_verb', 'base_verb', 'prefix'])
    print(f"После удаления дубликатов: {len(merged_df_unique)} строк")
    
    # Сохранить объединённый файл
    output_file = os.path.join(OUT_DIR, "serbian_verb_final_hf_wiki_merged.csv")
    merged_df_unique.to_csv(output_file, index=False, encoding="utf-8-sig")
    print(f"\n✓ Объединённый файл сохранён: {output_file}")
else:
    print("Файлы не найдены")
```

## 4) Display Sample of Merged Data

```python
# Показать первые строки объединённого датасета
if files and 'merged_df_unique' in locals():
    print("Первые 10 строк объединённого датасета:")
    print(merged_df_unique.head(10))
    print("\nСтатистика по префиксам:")
    print(merged_df_unique['prefix'].value_counts())
```
